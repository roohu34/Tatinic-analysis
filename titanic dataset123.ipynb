```python
x=a[["SibSp","Sex","Age","Pclass"]].replace(["male","female"],[0,1])
x
```

```python
y=a[["Survived"]]
y
```

```python
x.isnull().sum()
```

```python
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state =None, test_size = 0.2, shuffle = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
```

```python
KNN=KNeighborsClassifier()
```

```python
KNN.fit(xtrain,ytrain)
```

```python
KNN.predict(xtest)
```

```python
KNN.score(xtrain,ytrain)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.predict(xtest)
```

```python
clf.score(xtrain,ytrain)
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtrain,ytrain)
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC(degree=9,gamma="auto",random_state=None,kernel="linear",shrinking=True,tol=2)
```

```python
svm.fit(xtrain,ytrain)
```

```python
svm.predict(xtest)
```

```python
svm.score(xtrain,ytrain)
```
